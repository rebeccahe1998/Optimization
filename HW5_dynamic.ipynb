```
import numpy as np
```

```
# time needed with additional money
M = np.array([
[0,15,20,18,25],
[5,12,16,15,21],
[10,9,13,12,18],
[15,8,11,10,16],
[20,7,9,9,14],
[25,6,8,8,12],
[30,5,7,7,11],
[35,4,7,6,10]
])
n=4    # number of projects
m=len(M)    # number of budgests
```

```
V=np.zeros((m,n))  # for value of total completion time given budget
p = np.zeros((m,n)) # the optimal decision of budgests given to 4 projects given remaining budget
V[:,n-1]=M[:,n]  # given budgets, the toal completion time if can only do the last project
p[:,n-1] = range(m)  # optimal value
print(M)
print(V)
```

```
for t in range(n-2,-1,-1):
    
    for s in range(m): #[0,7]
        #p[s,t]=0
        V[s,t] = M[0,t+1]+V[s,t+1]
        if s>0:
            for k in range(1,s+1):  # loop over budgets [1,7] 
                if(V[s,t] > M[k,t+1] +V[int(s-k),t+1] ):  # total completion time of t and after with budget s 
                    V[s,t] = M[k,t+1] +V[int(s-k),t+1]    #  is min {time for project t given $k  and projects after t given $s-k}
                    
                    p[s,t]=k
s=7
x=[]
for i in range(4):
    x.append(p[int(s),i])
    s=s-p[int(s),i]
V
```
