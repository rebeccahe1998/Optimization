### Solutions to Problem 3
fatest path

```python
C = np.array([
     [np.Inf, 5, 3, 12, 9],
     [5, np.Inf, 7, 10, 2],
     [3, 7, np.Inf, 4, 7],
     [10, 9, 4, np.Inf, 1],
     [7, 2, 8, 1, np.Inf]
])
n = np.shape(C)[0]
```

```python
n
```

```python
V = np.zeros((n,n))
pi = np.zeros(n)
V[:,0] = C[:,0]
```

```python
s = 0
for t in range(1,n):
    for k in range(n):
        if (k != s):
            V[k,t] = V[k,t-1]
            for l in range(n):
                if (min([V[l,t-1],C[l,k]]) > V[k,t]):
                    V[k,t] = min([V[l,t-1],C[l,k]])
                    pi[k] = l
pi
```
